```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
import math

# Load the dataset
file_path = '/content/Daily_Public_Transport_Passenger_Journeys_by_Service_Type_20241129.csv'  # Replace with your actual file path
data = pd.read_csv(file_path)

# Preprocess the data
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)  # Convert Date to datetime
data = data.sort_values(by='Date')  # Sort by Date

# Service types to forecast
service_columns = ["Local Route", "Light Rail", "Peak Service", "Rapid Route", "School"]

# Generate dates for the next 7 days
forecast_dates = pd.date_range(start=data['Date'].iloc[-1] + pd.Timedelta(days=1), periods=7)

# Initialize dictionary to store forecasts
forecast_results = {}

# Normalize data using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

for column in service_columns:
    print(f"#### Forecasting for {column} using LSTM ####")
    try:
        # Select and scale the data
        ts_data = data[column].values.reshape(-1, 1)
        ts_data_scaled = scaler.fit_transform(ts_data)

        # Create a function to prepare data for supervised learning
        def create_dataset(data, time_step=1):
            X, y = [], []
            for i in range(len(data) - time_step - 1):
                X.append(data[i:(i + time_step), 0])
                y.append(data[i + time_step, 0])
            return np.array(X), np.array(y)

        # Prepare data for LSTM model
        time_step = 10  # Number of previous days to use for predicting the next day
        X, y = create_dataset(ts_data_scaled, time_step)

        # Reshape X to be 3D [samples, time steps, features]
        X = X.reshape(X.shape[0], X.shape[1], 1)

        # Split data into training and testing sets
        train_size = int(len(X) * 0.8)
        X_train, X_test = X[:train_size], X[train_size:]
        y_train, y_test = y[:train_size], y[train_size:]

        # Define the LSTM model
        model = Sequential()
        model.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], 1)))
        model.add(Dense(units=1))
        model.compile(optimizer='adam', loss='mean_squared_error')

        # Train the model
        model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=1)

        # Predict the next 7 days
        last_sequence = ts_data_scaled[-time_step:].reshape(1, time_step, 1)
        forecast_scaled = []
        for _ in range(7):
            predicted_value = model.predict(last_sequence)
            forecast_scaled.append(predicted_value[0][0])
            last_sequence = np.append(last_sequence[:, 1:, :], predicted_value.reshape(1, 1, 1), axis=1)

        # Inverse transform the predictions
        forecast = scaler.inverse_transform(np.array(forecast_scaled).reshape(-1, 1))

        # Store forecast results
        forecast_results[column] = forecast

        # Print forecast results
        print("| Date       | Predicted Passenger Journeys |")
        print("|------------|-------------------------------|")
        for date, value in zip(forecast_dates, forecast):
            print(f"| {date.date()} | {round(value[0])}                        |")
        print("\n")

    except Exception as e:
        print(f"Could not forecast {column} due to error: {e}\n")

# Plot the historical data and forecast using LSTM
for column in service_columns:
    if column in forecast_results:  # Ensure forecast exists for the column
        plt.figure(figsize=(10, 6))

        # Historical data
        plt.plot(data['Date'], data[column], label="Historical Data")

        # Forecast data from LSTM
        forecast = forecast_results[column]
        plt.plot(forecast_dates, forecast, label="Forecast (LSTM)", linestyle='--', marker='o')

        # Plot details
        plt.title(f"{column} - Passenger Journeys Forecast (LSTM)", fontsize=14)
        plt.xlabel("Date", fontsize=12)
        plt.ylabel("Passenger Journeys", fontsize=12)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
file_path = '/content/Daily_Public_Transport_Passenger_Journeys_by_Service_Type_20241129.csv'  # Replace with your actual file path
data = pd.read_csv(file_path)

# Preprocess the data
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True)  # Convert Date to datetime
data = data.sort_values(by='Date')  # Sort by Date

# Service types to forecast
service_columns = ["Local Route", "Light Rail", "Peak Service", "Rapid Route", "School"]

# Display the first few rows to check the data
data.head()
```

```python
# Remove outliers using the IQR method
for column in service_columns:
    print(f"#### Removing outliers for {column} ####")
    try:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = data[column].quantile(0.25)
        Q3 = data[column].quantile(0.75)
        IQR = Q3 - Q1

        # Define the lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Remove outliers
        data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

        print(f"Outliers removed for {column}. Remaining rows: {data.shape[0]}\n")

        # Optional: You can plot the boxplot to see the distribution after removing outliers
        plt.figure(figsize=(8, 6))
        plt.boxplot(data[column], vert=False)
        plt.title(f"{column} - Boxplot after removing outliers")
        plt.show()

    except Exception as e:
        print(f"Error removing outliers for {column}: {e}\n")

# Display the cleaned data
data.head()
```

```python
# Normalize data using MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Generate dates for the next 7 days
forecast_dates = pd.date_range(start=data['Date'].iloc[-1] + pd.Timedelta(days=1), periods=7)

# Prepare data for LSTM model
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Prepare data for LSTM model for one service column, e.g., "Local Route"
column = "Local Route"
ts_data = data[column].values.reshape(-1, 1)
ts_data_scaled = scaler.fit_transform(ts_data)

# Prepare dataset for LSTM
time_step = 10  # Number of previous days to use for predicting the next day
X, y = create_dataset(ts_data_scaled, time_step)

# Reshape X to be 3D [samples, time steps, features]
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Display prepared data dimensions
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Define the LSTM model
model = Sequential()
model.add(LSTM(units=25, return_sequences=False, input_shape=(X_train.shape[1], 1)))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=1)

# Evaluate the model
train_score = model.evaluate(X_train, y_train, verbose=0)
test_score = model.evaluate(X_test, y_test, verbose=0)

train_score, test_score
```
